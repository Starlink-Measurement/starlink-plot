```python
# Requirements:
# Three files: delay.csv, bandwidth-tcp.csv, and bandwidth-udp.csv
LOC_SAMI_BURNABY = 0
LOC_AWS_N_CALIFORNIA = 'N. California'
LOC_AWS_TOKYO = 'Tokyo'
LOC_AWS_LONDON = 'London'
LOC_AWS_SYDNEY = 'Sydney'
LOC_AWS_SINGAPORE = 'Singapore'
LOC_AWS_SAO_PAULO = 'Sao Paulo'
LOC_AWS_BAHRAIN = 'Bahrain'
LOC_AWS_MUMBAI = 'Mumbai'
LOC_AWS_CAPE_TOWN = 'Cape Town'
region_id_mapping = {LOC_AWS_N_CALIFORNIA: 100, LOC_AWS_TOKYO: 101, LOC_AWS_LONDON: 102, LOC_AWS_SYDNEY: 103, LOC_AWS_SINGAPORE: 104, LOC_AWS_SAO_PAULO: 105, 
                     LOC_AWS_BAHRAIN: 106, LOC_AWS_MUMBAI: 107, LOC_AWS_CAPE_TOWN: 108}
```

```python
def format_bandwidth_df(df):
    df_bandwidth = df.copy()[['Timestamp', 'bandwidth', 'bytes', 'Regions']]
    df_bandwidth = df_bandwidth.rename(columns={'bandwidth': 'avg. bandwidth', 'bytes': 'avg. transfer bytes', 'Regions': 'destination'})
    df_bandwidth['source'] = LOC_SAMI_BURNABY
    df_bandwidth['destination'] = df_bandwidth['destination'].map(lambda x: region_id_mapping[x])
    return df_bandwidth
```

```python
df_bandwidth_tcp = format_bandwidth_df(df[df['Protocol'] == 'TCP'])
df_bandwidth_udp = format_bandwidth_df(df[df['Protocol'] == 'UDP'])
```

```python
df_delay = df_send.groupby(['Regions', 'Timestamp'])['rtt'].agg(['mean', 'max', 'min']).reset_index()
df_delay = df_delay.rename(columns={'mean': 'avg. rtt', 'max': 'max. rtt', 'min': 'min. rtt', 'Regions': 'destination'})
df_delay['source'] = LOC_SAMI_BURNABY
df_delay['destination'] = df_delay['destination'].map(lambda x: region_id_mapping[x])
df_delay
```

```python
df_delay.dtypes
```

```python
#pd.to_datetime(df_delay.loc[1:5, 'Timestamp']).astype(int) / 1e6
df_delay['Timestamp'] = pd.to_datetime(df_delay.loc[:, 'Timestamp']).astype(int) / 1e6
df_bandwidth_tcp['Timestamp'] = pd.to_datetime(df_bandwidth_tcp['Timestamp']).astype(int) / 1e6
df_bandwidth_udp['Timestamp'] = pd.to_datetime(df_bandwidth_udp['Timestamp']).astype(int) / 1e6
print(df_delay)
print(df_bandwidth_tcp)
print(df_bandwidth_udp)
```

```python
df_delay[['source', 'destination', 'min. rtt', 'max. rtt', 'avg. rtt']].to_csv('output/GNN/delay.csv', index=False)
df_bandwidth_tcp[['source', 'destination', 'avg. bandwidth', 'avg. transfer bytes']].to_csv('output/GNN/bandwidth_tcp.csv', index=False)
df_bandwidth_udp[['source', 'destination', 'avg. bandwidth', 'avg. transfer bytes']].to_csv('output/GNN/bandwidth_udp.csv', index=False)
```

```python
df_delay[['source', 'destination', 'min. rtt', 'max. rtt', 'avg. rtt', 'Timestamp']].to_csv('output/GNN/dt_delay.csv', index=False)
df_bandwidth_tcp[['source', 'destination', 'avg. bandwidth', 'avg. transfer bytes', 'Timestamp']].to_csv('output/GNN/dt_bandwidth_tcp.csv', index=False)
df_bandwidth_udp[['source', 'destination', 'avg. bandwidth', 'avg. transfer bytes', 'Timestamp']].to_csv('output/GNN/dt_bandwidth_udp.csv', index=False)
```
